# Concatenate Data Frames

```python
import pandas as pd
```

```python
india_weather = pd.DataFrame({
    "city": ["Hyderabad","Banglore","Pune"],
    "temperature":[33,28,40],
    "humidity":[80,60,78]
})
```

```python
india_weather
```

```python
US_weather = pd.DataFrame({
    "city": ["new york","washington","arizona"],
    "temperature":[3,2,40],
    "humidity":[20,70,38]
})
```

```python
US_weather
```

```python
df = pd.concat([india_weather,US_weather],ignore_index=True)
df
```

```python
#row wise concatenation
df = pd.concat([india_weather,US_weather],axis=1,ignore_index=True)
df
```

# Merge Data Frames like Join in SQL


```python
temp_df = pd.DataFrame({
    "city": ["Hyderabad","Banglore","Delhi"],
    "temperature":[33,28,40]
})
```

```python
temp_df
```

```python
humidity_df = pd.DataFrame({
     "city": ["Hyderabad","Banglore","Delhi","Kerala"],
    "humidity":[63,78,40,12]
})
humidity_df
```

```python
#merge only works when the city name matches in both dfs and merged corresponding to it suppose for hybd 
#it found temperature 33 in tempdf and 63 humid for hyd and merged based on the city this is also can be said as
#inner join

df = pd.merge(temp_df,humidity_df,on='city')
df
```

# Outer Join

```python
df = pd.merge(temp_df,humidity_df,on='city',how='outer')
df
```

```python
#compares based on left else gives whether it has cities in both the df
df = pd.merge(temp_df,humidity_df,on='city',
              how='left')
df
```

```python
df = pd.merge(temp_df,humidity_df,on='city',
              how='right')
df
```
