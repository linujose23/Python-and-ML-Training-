# MatplotLib or Pyplot
Its based on Matlab. Each pyplot function makes some changes to figure like creates fig, creates plot area, decorates with lines and labels...

```python
import matplotlib.pyplot as plt
```

```python
plt.xlabel("indices")
plt.ylabel("numbers")
plt.plot([2,3,4,5,6,7])
```

```python
plt.xlabel("indices")   #by default it takes indexes at x
plt.ylabel("numbers")  #and y we are giving below
plt.title("myplot")
plt.plot([2,3,4,5,6,7,4])
```

```python
plt.xlabel("indices")  
plt.ylabel("numbers")  
plt.title("myplot")
plt.grid()      #for plotting it in a grid
plt.plot([2,3,4,5,6,7,4],[1,2,5,7,10,12,16])    #takes x and y x is the first list and y is the second one
```

```python
plt.xlabel("indices")  
plt.ylabel("numbers")  
plt.title("myplot")
plt.grid()      #for plotting it in a grid
plt.plot([2,3,4,5,6,7,4],[1,2,5,7,10,12,16],'ro')    #ro here is red and o the shape
```

```python
import numpy as np
#start end step size
t = np.arange(0.,5.,0.2)

#x-axis, y-axis, 'color Shape',label
#b-- ->blue dashed line
#rs ->red Squares
#g^ ->green triangles
#yo ->yellow circles
#here its taking x and y x is t and y is t** or t square
plt.plot(t,t**2,'b--',label='^2');
plt.plot(t,t**2.2,'rs',label='2.2');
plt.plot(t,t**2.5,'g^',label='2.5');
plt.plot(t,t**2.8,'yo',label='2.8');
plt.legend() #to show symbol info legend is used to explain about the data
plt.grid()
plt.plot()
```

```python
x = [1,2,3,4]
y = list(map(lambda i:i**2,x))
```

```python
plt.plot(x,y,linewidth=5)
```

```python
x = [1,2,3,4]

y = [1,4,9,16]

colorShape = "b--";   #b -- here is blue dashed line

plt.plot(x,y,colorShape,linewidth=5.0)
```

```python
x1 = [1,2,3,4]
y1 = [1,4,9,16]
x2 = [1,2,3,4]
y2 = [2,4,6,8]

#it returns 2 lines ->lines[0],lines[1]
lines = plt.plot(x1,y1,x2,y2)

#using setProperty method we can configure line
#its keyword args style of parameters
plt.setp(lines[0],color='b',linewidth=2.0)    #here it takes x1 and y1 in line[0]

#its matlab style of args
plt.setp(lines[1],'color','g',linewidth=5)   #here it takes x2 and y2 in line[1]

plt.grid()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
#just a simple function above with xyz formula just to show its plot
# start ,end axis and step size
t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.3)

plt.figure(1)    #this creates a figure for us 

# 211 indicates 2->2rows, 1->1col, 1->1st subplot
plt.subplot(211)                        #2 rows and 1 column and 1st sublot is 211 here
plt.grid()                        #just for adding grids to the figure

#x x axis, y axis, colorShape
plt.plot(t1,f(t1), 'b--',label='t1')          #plotting t1 and the f function output after passing t1
plt.legend() #line info at cornor

plt.subplot(212)                             #2 rows and 1 column and 2nd sublot is 212 here
plt.grid()
plt.plot(t2,f(t2),'g^',label='t2')            #plotting t2 and the f function output after passing t2
plt.legend()  #legend plot is the plot on top right part which shows dased blue and green line pattern

plt.show()
```
